```
import pandas as pd
import numpy as np
pd.__version__
```

# Numpy

```
# Cambiamos el número de decimales 
pd.options.display.float_format = '{:.2f}'.format
```

```
# Cambiamos el número de decimales 
np.set_printoptions(precision=1)
```

```
# Creamos arreglos numpy con datos aleatorios
x1 = np.random.rand(2,5)*10
x1
```

```
x2 = np.random.rand(2,5)*-1
x2
```

```
# Concatenamos 2 arrglos de numpy
np.concatenate([x1,x2])
```

```
# Vemos la dimenciones de arreglo concatenado
np.concatenate([x1,x2]).shape
```

```
# Cambiamos las dimensiones del arreglo concatenado
np.concatenate([x1,x2], axis = 1)
```

```
# Vemos las dimensiones que tiene el arreglo concatenado
np.concatenate([x1,x2], axis = 1).shape
```

# PANDAS

```
# Creamos series con un cierto indice y con datos de arrglo de numpy
s1 = pd.Series(x1[0],index = ['a','b','c','d','e'])
s1
```

```
s2 = pd.Series(x2[0],index = ['c','d','e','f','g'])
s2
```

```
# Conatenamos 2 series de pandas de forma de columnas
pd.concat([s1,s2], axis=1)
```

```
# Reiniciamos el indice (Colocando como columna)
s1.reset_index()
```

```
# Reiniciamos el indice (Borrandolo)
s1.reset_index(drop=True)
```

```
# Concatenamos las series borrando sus indices y colocandolo como columnas 
pd.concat([s1.reset_index(drop=True),s2.reset_index(drop=True)], axis=1)
```

```
# Creamos un nuevo DF con valores aletorios y en un cierto shape con las columnas a y b
df1 = pd.DataFrame(np.random.rand(3,2)*10,columns=['a','b'])
df1
```

```
# Creamos un DF igual al anterior pero con un indice de 2,3,4
df2 = pd.DataFrame(np.random.rand(3,2)*-1,columns=['a','b'], index = [2,3,4])
df2
```

```
# Concatenamos 2 DF
pd.concat([df1,df2])
```

```
# Concatenamos 2 DF y los colocamos como columnas
pd.concat([df1,df2], axis=1)
```

```
# Concatenamos 2 DF, en columnas y obtenemos el ejemplar que se repite en ambos DF.
pd.concat([df1,df2], axis=1, join='inner')
```

```
# Concatenamos y reiniciamos indices borradolos
pd.concat([
           df1.reset_index(drop=True),
           df2.reset_index(drop=True)
           ], axis=1)
```

```
# Le agregamos 2 veces el mismo DF a df1
df1.append(df2).append(df2)
```

```
# Cremos una transpues de las columnas x las filas de un DF
df1.T
```

```
# A la transpues de un DF se le agrega otro DF en transpuesta y eso se coloca en transpuesta
df1.T.append(df2.T).T
```
