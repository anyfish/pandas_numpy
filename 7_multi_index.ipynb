```python
df_pob = pd.read_csv('population.csv', skiprows=4)
df_pob
```

```python
df_pob.columns
```

```python
df_mod = df_pob[['Country Name', '2015', '2016', '2017', '2018']]
df_mod
```

```python
# melt .:. derretir
df_mod = df_mod.melt(id_vars=['Country Name'], value_vars=['2015', '2016', '2017', '2018'], var_name='Year', value_name='Population')
df_mod
```

```python
df = df_mod
```

```python
df
```

```python
# Cambiar la categoria de una columnas en especifico
df['Year'] = pd.Categorical(df['Year'].apply(str))
df.dtypes
```

```python
# Buscar si datos esta en un columna
# Este puede ser un filtro que apliquemos al DF completo
idx_filtro = df['Country Name'].isin(['Aruba','Colombia'])
idx_filtro
```

```python
# Uso del filtro para el DF completo
df_sample = df[idx_filtro]
df_sample
```

```python
df_sample = df_sample.set_index(['Country Name', 'Year']).sort_index()
df_sample
```

```python
# Seleccion de un indice
df_sample.loc['Colombia',:]
```

```python
# Selección con Multiples indices (multi index)
df_sample.loc['Colombia',:].loc['2016']
```

```python
# Ordenar de difentes formas un multi index
df_countries = df.set_index(['Country Name', 'Year']).sort_index(ascending=[True, True])
df_countries
```

```python
# Encontrar por columna una doble rango especifico y ordenarlo de forma acendente
# Se tiene que instanciar un metodo
ids = pd.IndexSlice
df_countries.loc[ids['Aruba':'Austria','2015':'2017'],:].sort_index()
```

```python
# En un multi index obtener los indices principales
df_countries.index.get_level_values(0)
```

```python
# En un multi index obtener los indices secundarios
df_countries.index.get_level_values(1)
```

```python
# Entrar a difentes niveles de indices
df_countries['Population']['Colombia']['2018']
```

```python
# Operación Aritmetica en las diferentes categorias por columna
df_countries.sum(level='Year')
```

```python
# Convertir un indice en columnas
df_sample.unstack('Year')
```

```python
df_sample.unstack('Country Name')
```
